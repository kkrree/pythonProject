```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.DataFrame(boston.target, columns=['MEDV'])
df = pd.concat([X,y], axis=1) #가로 방향 연결
df.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#데이터 셋 분할 (학습용:검증용=7:3)
#random_state : 난수 발생을 위한 seed의 인자값
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
model = LinearRegression().fit(X_train, y_train)
```

```python
print('학습용:', model.score(X_train,y_train))
print('검증용:', model.score(X_test,y_test))
```

```python
#상관계수
coef = model.coef_

#절편
intercept = model.intercept_

print(coef)
print('==========')
print(intercept)
```

```python
model.feature_names_in_
```

```python
#범죄율이 한단위 증가하면 집값은 약 153 달러 하락한다. >crim 
#찰스강의 경계에 위치한 경우 집값은 약 1622 달러 상승한다. >chas
#방갯수가 한단위 증가하면 집값은 약 3352 달러 상승한다. >rm
#재산세율이 한단위 증가하면 집값은 약 1223 달러 하락한다. >tax
```

```python
pred = model.predict(boston.data)

plt.scatter(boston.target, pred)
plt.xlabel('real house price')
plt.ylabel('predicted house price')
plt.title('compare of predicted house price')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

dfX = pd.DataFrame(boston.data, columns=boston.feature_names)
dfy = pd.DataFrame(boston.target, columns=['MEVD'])

model_boston2 = sm.OLS(dfy, dfX)
result_boston2 = model_boston2.fit()

print('회귀계수\n', result_boston2.params) #기울기
print('==========')
print('R-Squared\n', result_boston2.rsquared) #회귀모델의 설명력 95.8%의 설명력
print('==========')
print('P values\n', result_boston2.pvalues)
```

```python
#요약 결과 출력
print(result_boston2.summary)
```

- 전진선택법 : 공집합인 상태에서 변수를 1개씩 추가하면서 전체 모델의 성능이 좋아질 때까지 변수를 계속 증가시키는 방법
- 후진제거법 : 전체 변수를 대상으로 미리 정의한 모든 통계 수치(유의수준, 다중공선성 등)를 만족할 때까지 반복적으로 변수를 하나씩 제거하는 방법(pvalue가 더 중요함)
   
변수들 중 탈락시킬 변수 선정   
['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']   
pvalue가 0.05보다 큰 값을 탈락시킨다.

```python
print('P values\n', result_boston2.pvalues)
```

INDUS, NOX, AGE 탈락

```python
#다중공선성은 VIF값으로 확인할 수 있다.
#다른 변수에 의족적일수록 VIF가 커진다(다중공선성)
#독립변수가 서로 의존하게 되면 과적합화 문제가 발생하여 모델의 안정성이 떨어질 수 있다.
#VIF(Variance Inflation Factor) 분산 팽창 요인

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(dfX.values, i) for i in range(dfX.shape[1])]
vif['변수'] = dfX.columns
vif = vif.sort_values('VIF').reset_index(drop=True)
vif
#vif값은 작을수록 좋은대 5보다 큰 값은 다중공선성이 큰 변수라고 볼 수 있다.
```

```python
#VIF가 높은 값들을 제거하기 전의 모형
import numpy as np
model_boston1 = sm.OLS(np.log(dfy),dfX)
result_boston1 = model_boston1.fit()
print(result_boston1.summary())
```

```python
#VIF가 높은 변수들을 제거하고 최종적으로 선택한 독립변수들로 만든 모형
cols = ['intercept', 'CHAS', 'scale(B)', 'scale(CRIM)', 'scale(AGE)', 'scale(np.log(PTRATIO))', 'scale(RM)', 'scale(np.log(LSTAT))'] 

model_boston2 = sm.OLS(np.log(dfy), dfX[cols])
result_boston2 = model_boston2.fit()
print(result_boston2.summary())
```
